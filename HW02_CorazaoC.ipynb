Section 5c: Create a Barplot based on Location Condition 
 
• Create a code block named df_locCond that is a groupby of ‘Location Condition’ 
that takes a count of ‘Sale Price’. 
o Reset the index 
o Sort df_locCond from largest to smallest.  
o To change df_locCond, make sure to start the line of code with: 
df_locCond =  

```python
# CodeBlock Section 5c1
df_locCond = df_realestate.groupby('Location Condition')['Sale Price'].count().reset_index()
df_locCond = df_locCond.sort_values('Sale Price', ascending = False)
df_locCond
```

• Create a code block that creates a variable named var_total that sums up all of 
values from df_locCond[‘Sale Price’].   
o var_total = df_locCond[‘Sale Price’].sum()

```python
# CodeBlock Section 5c2
var_total = df_locCond['Sale Price'].sum()
var_total
```

• Create a code block that creates a new column for df_locCond named ‘Percent’ 
o The new column ‘Percent’ is df_locCond[‘Sale Price’] divided by var_total. 

```python
# CodeBlock Section 5c3
df_locCond['Percent'] = df_locCond['Sale Price'] / var_total
df_locCond['Percent']
```

• Create a code block that creates a barplot for Location Condition where: 
o Data is df_LocCond 
o x is ‘Location Condition’  
o y is ‘Percent’ 
o Other properties, including labels are similar to the barplot from Section 5b. 

```python
# CodeBlock Section 5c4
sns.barplot(x= 'Location Condition', y= 'Percent', data = df_locCond, color = 'b')
plt.title('Percentage of Houses by Location Condition', fontweight = 'bold', fontsize = '18', horizontalalignment = 'center')
plt.xlabel('Location Condition', fontweight = 'bold', fontsize = '14', horizontalalignment = 'center')
plt.ylabel('Percentage', fontweight = 'bold', fontsize = '14', horizontalalignment = 'center')
```

# CodeBlock # CodeBlock Section 5c5
Based on the barplot above the normal location condition has the highest percentage in sale price by up to 0.8. The data is almost exclusively normal while the other location conditions have a percentage of less than 0.2. We can conclude that people are willing to spending more in a normal location condition.

Section 6c: Create Z-Scores for Sale Price and Violinplot  
• Create a code block to create two variables: 
o  One for the mean ‘Sale Price’ and name it mean_price. 
o One for the standard deviation of ‘Sale Price’ and name it stdev_price. 
o Print the mean and standard deviation for the output of this code block.  
Similar to below.

```python
# CodeBlock Section 6c1
mean_price = df_realestate['Sale Price'].mean()
stdev_price = df_realestate['Sale Price'].std()
print('The mean sale price is $',mean_price)
print('The Standard deviation of sale price is $',stdev_price)
```

• Create a code block that creates a z-score for ‘Sale Price’. 
o Name the new column ‘z-score’ 
o The new column will be in df_realestate 
o Use mean_price and stdev_price to create a z-score column.  #See 
C1.S3.Py09 to calculate the z-score.

```python
# CodeBlock Section 6c2
mean_price = df_realestate['Sale Price'].mean()
stdev_price = df_realestate['Sale Price'].std()
df_realestate['z_score'] = (df_realestate['Sale Price'])-mean_price/stdev_price
df_realestate
```

• Create a code block to create a subplot with two Violinplots. 
o Use - plt.subplot(121) to indicate the first violinplot for raw data 
• y is "Sale Price" 
• x is "Land Contour" 
• data is df_realestate 
• color is 'b' 
• Add a title and x and y labels  
o Use - plt.subplot(122) to indicate the second violin plot for z-scores 
• y is "z-score" 
• x is "Land Contour" 
• data is df_realestate 
• color is 'g' 
• Add a title and x and y labels

```python
# CodeBlock Section 6c3

plt.subplot(121)

sns.violinplot(y = "Sale Price", x = "Land Contour", data = df_realestate, palette = "Blues")
plt.title('Sale Price by Land Contour', fontweight = 'bold', fontsize = '18', horizontalalignment = 'center')
plt.xlabel('Land Contour', fontweight = 'bold', fontsize = '14', horizontalalignment = 'center')
plt.ylabel('Sale Price', fontweight = 'bold', fontsize = '14', horizontalalignment = 'center')

plt.subplot(122)

sns.violinplot(y = "z_score", x = "Land Contour", data = df_realestate, palette = "Greens")
plt.title('z_score by Land Contour', fontweight = 'bold', fontsize = '18', horizontalalignment = 'center')
plt.xlabel('Land Contour', fontweight = 'bold', fontsize = '14', horizontalalignment = 'center')
plt.ylabel('z_score', fontweight = 'bold', fontsize = '14', horizontalalignment = 'center')
```
